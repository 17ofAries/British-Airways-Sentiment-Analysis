```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import nltk
```

```python
#Download nltk resources

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df = pd.read_csv('skytrax_reviews.csv')
df.head()
```

```python
df['Content'] = df['Content'].str.replace('Not Verified', '', regex=False).str.replace('Trip Verified', '', regex=False)
df['Seat Type'] = df['Seat Type'].str.replace(' Class', '', regex=False)
df.head()
```

```python
def cleanText(text):
    text = text.lower() #Convert to lower case
    text = re.sub(r'[^A-Za-z\s]', '', text) #Remove punctuations and special characters
    words = word_tokenize(text) #Tokenize text
    words = [word for word in words if word not in stopwords.words('english')] #Remove stopwords
    
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words] #Lemmatize the words
    
    clean_text = ' '.join(words)
    return clean_text
```

```python
# Apply cleanText function to Content column
df['Cleaned Content'] = df['Content'].apply(cleanText)
df['Cleaned Title'] = df['Title'].apply(cleanText)
```

```python
df.head()
```

```python
df.drop(columns=['Title','Content'], axis=1, inplace=True)
df.head()
```

```python
df['Text'] = df['Cleaned Title'] + ' ' + df['Cleaned Content']
df.drop(columns=['Cleaned Content','Cleaned Title'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df['Day of Month'] = df['Date'].dt.day
df['Day of Week'] = df['Date'].dt.day_name()

df.head()
```

```python
print("Duplicate Reviews:", df.duplicated().sum())
```

```python
df['Text Length'] = df['Text'].apply(len)
df.head()
```

```python
df.to_csv('cleand_skytrax_reviews.csv', index=False)
print('Data Cleaning completed and saved to cleaned_skytrax_reviews.csv')
```
